```python
import pandas as pd

hitters = pd.read_csv("Hitters.csv")
hitters.rename(columns={'Unnamed: 0': 'Player'}, inplace=True)
hitters
```

### Drop samples with missing target

```python
import numpy as np
missing_y = hitters['Salary'].index[hitters['Salary'].apply(np.isnan)]
missing_y
```

```python
hitters.drop(missing_y, inplace=True)
```

```python
X = hitters.drop(['Salary'], axis=1)
y = hitters['Salary']
print(X.shape, y.shape)
```

### Preprocess data

```python
numeric_features = X._get_numeric_data().columns.tolist()
categorical_features = [feature for feature in X.columns.tolist() if feature not in numeric_features]
print("Numeric features:\n", numeric_features, '\n\nCategorical features:\n', categorical_features)
```

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import KNNImputer, SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer

numeric_transformer = Pipeline(steps=[
    ('imputer', KNNImputer(n_neighbors=5, weights='distance')),
    ('scaler', StandardScaler())])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent', fill_value='missing')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)])

X_preprocessed = preprocessor.fit_transform(X)
```

```python
# Both categorical and numeric features
X_preprocessed = pd.DataFrame(X_preprocessed.toarray())
X_preprocessed
```

```python
# Only numeric features
X_preprocessed_numeric = pd.DataFrame(numeric_transformer.fit_transform(X._get_numeric_data()))
X_preprocessed_numeric.columns = numeric_features
X_preprocessed_numeric
```

# 1.1 Lasso regularization

### Predict Salary with the Lasso

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_preprocessed_numeric, y, test_size = 0.2, random_state=42)
```

```python
from sklearn.linear_model import Lasso
ls1 = Lasso(alpha=1)
ls1.fit(X_train, y_train)
print("Training R2 score:", ls1.score(X_train, y_train))
print("Testing R2 score:", ls1.score(X_test, y_test))
```

### Lasso cross validation & parameter tuning

```python
from sklearn.model_selection import cross_val_score
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

alphas = np.logspace(-10, 1, 100)
scores = [cross_val_score(Lasso(alpha), X_preprocessed_numeric.values, y, cv=5).mean() for alpha in alphas]

plt.figure(figsize=(6,4))
plt.plot(alphas, scores, label=Lasso.__name__)
plt.legend(loc='upper right')
plt.title("R2 score on the holdout dataset")
plt.xlabel('alpha')
plt.ylabel('cross validation score')
plt.tight_layout()
plt.show()
```

```python
print("The highest R^2 score on the test data is %0.3f" %max(scores))
print("The best regularization pentalty is %0.3f" %alphas[np.argmax(scores)])
```

# 1.2 Ridge regularization

### Predict Salary with Ridge Regularization

```python
from sklearn.linear_model import Ridge
rg1 = Ridge()
rg1.fit(X_train, y_train)
print("Training R2 score:", rg1.score(X_train, y_train))
print("Testing R2 score:", rg1.score(X_test, y_test))
```

### Ridge coefficient path

```python
alphas = np.logspace(-10, 5, 100)
coefs = []
for a in alphas:
    ridge = Ridge(alpha=a, fit_intercept=False)
    ridge.fit(X_preprocessed_numeric, y)
    coefs.append(ridge.coef_)

xx = np.sum(np.abs(coefs), axis=1)
xx /= xx[-1]

plt.figure(figsize=(6,4))
plt.plot(xx, coefs)
ymin, ymax = plt.ylim()
#plt.vline(xx, ymin, ymax, linestyle='dashed')
plt.xlabel('|coef| / max|coef|')
plt.ylabel('Coefficients')
plt.title('Ridge Path')
plt.axis('tight')
plt.show()
```

```python
alphas = np.logspace(-10, 5, 100)
scores = [cross_val_score(Ridge(alpha), X_preprocessed_numeric.values, y, cv=5).mean() for alpha in alphas]

plt.figure(figsize=(6,4))
plt.plot(alphas, scores, label=Ridge.__name__)
plt.legend(loc='upper right')
plt.title("R2 on the holdout dataset")
plt.xlabel('alpha')
plt.ylabel('cross validation score')
plt.tight_layout()
plt.show()
```

```python
print("The highest R^2 score on the test data is %0.3f" %max(scores))
print("The best regularization pentalty is %0.3f" %alphas[np.argmax(scores)])
```

# 2 Bias-variance tradeoff

**Bias-variance tradeoff:** Bias refers to how the predicted values differ from the true values. Models with high bias are oversimplified and has high erros on both training and test data. Variance refers to how much the predicted value vary on different subsets of the data. Models with high variance pays too much attention to the training data (overfits) and is not generalizable on new unseen data. Bias-variance tradeoff refers finding a balanced complexity of the model so that it yields similar prediction error on the training and testing data. 

**Regularization:** Regularization prevents overfitting. It penalizes models that are more complicated by adding a penalty term that increases as the number of parameters increases. In problem 1, the naive lasso and ridge models use the default penalty parameter alpha=1.0. Comparing this to the optimal alphas (4.642 for Lasso and 162.975 for Ridge) obtained during cross validation, the naive models are overfitting on the training data and is not generalizable, a.k.a. low bias, high variance. Models using the optimal penalty creates a balance between bias and variance, and yields similar prediction scores/errors on the training and testing data.
